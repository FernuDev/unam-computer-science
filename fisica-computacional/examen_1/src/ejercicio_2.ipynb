## Exercise 2: Polynomial Function Visualization

**Objective**: Generate and visualize quadratic and cubic functions over a defined domain.

**Requirements**:
- Create vector **x** with evenly spaced values from -10 to 10 (step = 0.5)
- Define functions: f₁(x) = x² (quadratic) and f₂(x) = x³ (cubic)
- Build DataFrame with columns: x, f₁(x), f₂(x)
- Plot both functions using DataFrame data

**Mathematical Context**:
- f₁(x) = x²: Parabola with even symmetry, minimum at origin
- f₂(x) = x³: Cubic curve with odd symmetry, inflection point at origin

```python
"""
Polynomial Function Visualization - Exercise 2

Generates and visualizes quadratic and cubic polynomial functions
to analyze their behavior and compare growth rates.

Author: Luis Fernando Núñez Rangel
Institution: Facultad de Ciencias, UNAM
"""

import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f1(x: float) -> float:
    """
    Quadratic function f₁(x) = x².
    
    Args:
        x (float): Input value
        
    Returns:
        float: Square of input (x²)
        
    Properties:
        - Even function: f(-x) = f(x)
        - Parabola opening upward
        - Vertex at origin (0, 0)
        - Domain: (-∞, ∞)
        - Range: [0, ∞)
    """
    return x**2
```

```python
def f2(x: float) -> float:
    """
    Cubic function f₂(x) = x³.
    
    Args:
        x (float): Input value
        
    Returns:
        float: Cube of input (x³)
        
    Properties:
        - Odd function: f(-x) = -f(x)
        - S-shaped curve (sigmoid-like)
        - Inflection point at origin (0, 0)
        - Domain: (-∞, ∞)
        - Range: (-∞, ∞)
    """
    return x**3
```

```python
# Generate domain vector x with evenly spaced points
# Range: [-10, 10] with step = 0.5
# Method: List comprehension for efficient generation
# Total points: 41 (from -20*0.5 to 20*0.5)
x = [i * (1/2) for i in range(-20, 21)]
```

```python
# Evaluate functions at all points in domain
# Use list comprehension for vectorized-style computation
f1_vals = [f1(i) for i in x]  # Quadratic values
f2_vals = [f2(i) for i in x]  # Cubic values
```

```python
# Create structured DataFrame for analysis
# Columns: independent variable (x) and dependent variables (f1, f2)
df = pd.DataFrame({
    "x": x, 
    "f1(x)": f1_vals,  # Quadratic function values
    "f2(x)": f2_vals   # Cubic function values
})
df
```

```python
# Visualization: Comparative plot of polynomial functions
# Shows growth rate differences between quadratic and cubic

plt.figure(figsize=(10, 6))

# Plot both functions with distinct styling
plt.plot(x, f1_vals, label="f₁(x) = x²", linewidth=2, color='blue')
plt.plot(x, f2_vals, label="f₂(x) = x³", linewidth=2, color='red')

# Configure plot aesthetics
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.title("Polynomial Functions: Quadratic vs Cubic", fontsize=14)

# Add horizontal and vertical axis lines at origin
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()
```
