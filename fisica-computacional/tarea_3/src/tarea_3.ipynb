```python
"""
Kinematic Analysis of Runners - Data Visualization Module

This notebook analyzes experimental running data to visualize position-time
and velocity-time relationships for multiple runners. The velocity data is
computed from position data using numerical differentiation (Fortran program).

Author: Luis Fernando Núñez Rangel
Institution: Facultad de Ciencias, UNAM
Course: Computational Physics
"""

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load experimental data for runner 1
# Data format: whitespace-separated values (distance, time)
# Distance in kilometers, time in minutes
corredor_1 = pd.read_csv("../data/datos1n.dat", sep='\\s+', names=["Distancia (km)", "Tiempo (m)"])
corredor_1
```

```python
# Load experimental data for runner 2
# Same format as runner 1 for comparative analysis
corredor_2 = pd.read_csv("../data/datos2n.dat", sep='\\s+', names=["Distancia (km)", "Tiempo (m)"])
corredor_2
```

```python
# Comparative position-time plot for both runners
# This visualization allows identification of:
# - Different running strategies (constant pace vs variable)
# - Relative performance between runners
# - Distance covered over time progression

plt.title("Position vs Time - Comparative Analysis")
plt.xlabel("Time (min)")
plt.ylabel("Distance (km)")

# Plot both runners on same axes for comparison
plt.plot(corredor_1["Distancia (km)"], corredor_1["Tiempo (m)"], 
         label="Runner 1", marker='o', markersize=3)
plt.plot(corredor_2["Distancia (km)"], corredor_2["Tiempo (m)"], 
         label="Runner 2", marker='s', markersize=3)

plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig("../img/Posición.jpg", dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Load velocity data computed by Fortran numerical differentiation
# File generated by Corredores.f90 using 3-point Lagrange method
# Columns: tiempo_min, distancia_km, v_km_per_min, v_m_per_s

vel_corredor_1 = pd.read_csv("../data/datos1n_velocidad.dat", sep='\\s+')
vel_corredor_1
```

```python
# Load velocity data for runner 2
# Processed with same numerical method for consistency
vel_corredor_2 = pd.read_csv("../data/datos2n_velocidad.dat", sep='\\s+')
vel_corredor_2
```

```python
# Velocity-time comparative visualization
# Key insights from this plot:
# - Acceleration/deceleration phases during race
# - Peak velocity and sustained speed capabilities
# - Fatigue patterns (velocity degradation over time)
# - Pacing strategy differences between runners

plt.figure(figsize=(10, 6))
plt.title("Velocity vs Time - Comparative Analysis")
plt.xlabel("Time (min)")
plt.ylabel("Velocity (km/min)")

# Plot velocity profiles with distinct styling
plt.plot(vel_corredor_1["tiempo_min"], vel_corredor_1["v_km_per_min"], 
         label="Runner 1", linewidth=2, alpha=0.8)
plt.plot(vel_corredor_2["tiempo_min"], vel_corredor_2["v_km_per_min"], 
         label="Runner 2", linewidth=2, alpha=0.8)

plt.legend(loc='best')
plt.grid(True, alpha=0.3)
plt.savefig("../img/Velocidad.jpg", dpi=150, bbox_inches='tight')
plt.show()
```
