```python
import pandas as pd


pd.set_option("display.max_rows", None)    # mostrar todas las filas
pd.set_option("display.max_columns", None) # mostrar todas las columnas
pd.set_option("display.width", None)       # no cortar líneas
pd.set_option("display.max_colwidth", None)

df= pd.read_csv("psi.dat",
                 delim_whitespace=True,  # separador: espacios o tabs
                 header=None,            # no hay encabezado en el archivo
                 names=["posicion", "Psi"]) # nombres de columnas

```

```python
import matplotlib.pyplot as plt

# Graficar
plt.figure(figsize=(8,5))
plt.plot(df["posicion"], df["Psi"], marker="o", linestyle="-", color="b", label="E_1")

# Etiquetas
plt.xlabel("posicion")
plt.ylabel("Psi")
plt.title("Gráfica posicon vs Psi")
plt.legend()
plt.grid(True)

plt.show()
```
